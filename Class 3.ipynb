```python
class Queue:
    def __init__(self):
        self.items = []

    def enqueue(self, value):
        self.items.insert(0, value)
    def dequeue(self):
        return self.items.pop()
    #nice to have methods
    def peek(self):
        return self.items[len(self.items)-1]
        
    def size(self):
        return len(self.items)
        
    def is_empty(self):
        return self.items == []
```

```python
#from scratch implementation of queue

class QueueII:
    class __Node:
        def __init__(self, data):
            self.data = data
            self.next = None
            
    def __init__(self):
        #should we track the rear, front or both?
        self.rear = None #front of the queue
        self.front = None #end of the queue

    def enqueue(self, value):
        #recieves a value and "enqueues" it at the rear
        new_node = self.__Node(value)
        if not self.rear:
            self.rear = self.front = new_node
        else:
            self.rear.next = new_node
    def dequeue(self):
        if not self.front:
            raise IndexError("queue is empty")
        value = self.front.data
        self.front = self.front.next
        return value
    def peek(self):
        return self.front.data   
        #return but does not remove the data at the front node

    def size(self):
        count = 0 
        current = self.front
        while current:
            current = current.next
            count += 0
        return count
        #return the total number of Nodes in the queue
    def is_empty(self):
        return self.front == None
        #return True when the Queue is empty, False otherwise
```

```python
class SinglyLinkedList:
    class __Node:
        def __init__(self, data):
            self.data = data
            self.next = None

    def __init__(self):
        self.head = None

    def __str__(self):
        out = "["
        current = self.head
        if current:
            out += "%s" % current.data
            current = current.next
            while current:
                out += ", %s" % current.data
                current = current.next
        out += "]"
        return out

    def append(self, value):
        new_node = self.__Node(value)
        current = self.head
        if not current:
            self.head = new_node
        else:
            while current.next:
                current = current.next
            current.next = new_node

    def insert(self, index, value):
        new_node = self.__Node(value)
        if not self.head:
            self.head = new_node
        else:
            if index == 0:
                old_head = self.head
                self.head = new_node
                new_node.next = old_head
            else:
                prev = None
                current = self.head
                count = 0
                while count != index and current.next:
                    prev = current 
                    current = current.next
                    count += 1
                if count == index:
                    prev.next = new_node
                    new_node.next = current
                else:
                    current.next = new_node

    def remove(self, value):
        found = False
        if self.head:
            prev = None
            current = self.head
            while current:
                if current.data == value:
                    found = True
                    break
                prev = current
                current = current.next
            if found:
                prev.next = current.next
            if not found:
                raise ValueError("Value Not Found")
    def index(self, value):
        current = self.head 
        index = 0
        while current:
            if current.data == value:
                return index
            current = current.next
            index += 1
        raise IndexError("value not in list")

    def __getitem__(self, index):
        if self.head:
            count = 0
            current = self.head
            while count != index and current:
                current = current.next
                count += 0
            if count == index:
                return current.data
        raise IndexError("Index Out of Bound")

    def __len__(self):
        count = 0
        current = self.head
        while current:
            current = current.next
            count += 1
        return count

    def is_empty(self):
        return self.head is None
```

```python
# create an instance of both lists
pylist = []
sll = SinglyLinkedList()

# when the list is empty:
print("Python list: %s" % str(pylist))
print("Our SLL: %s" % str(sll))

pylist.append(1)
sll.append(1)
print("Python list: %s" % str(pylist))
print("Our SLL: %s" % str(sll))

for number in range(2, 6):
    pylist.append(number)
    sll.append(number)

print("Python list: %s" % str(pylist))
print("Our SLL: %s" % str(sll))
```
